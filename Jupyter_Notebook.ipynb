## Using Qgrid to explore dataframes

```python
!pip install qgrid
!jupyter nbextension enable --py --sys-prefix widgetsnbextension
```

```python
# example using qgrid:


import pandas as pd
df=pd.read_csv('placement.csv')

# using qgrid to access the above dataframe
import qgrid
qgrid_widget = qgrid.show_grid(df, show_toolbar=True)
qgrid_widget
```

```python
# if widget is not showing, try the below code
!jupyter nbextension enable --py --sys-prefix qgrid
!jupyter nbextension enable --py --sys-prefix widgetsnbextension
```
